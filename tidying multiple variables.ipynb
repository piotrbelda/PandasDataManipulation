```python
import pandas as pd
import numpy as np
```

```python
weightlifting=pd.read_csv('data/weightlifting_men.csv')
weightlifting
```

```python
wl_melt=weightlifting.melt(id_vars='Weight Category',var_name='sex_age',value_name='Qual Total')
```

```python
wl_melt.head(7)
```

```python
# splitting with str accessor
```

```python
sex_age=wl_melt['sex_age'].str.split(expand=True)
sex_age.head()
```

```python
sex_age.columns=['Sex','Age Group']
sex_age.head()
```

```python
# accessing first char of string
```

```python
sex_age.Sex=sex_age.Sex.str[0]
```

```python
# concatenating to a final tidy form
```

```python
wl_cat_total=wl_melt[['Weight Category','Qual Total']]
wl_tidy=pd.concat([sex_age,wl_cat_total],axis=1)
wl_tidy.head()
```

```python
cities=pd.read_csv('data/texas_cities.csv')
cities
```

```python
# method without .str() accessor
```

```python
cols=['latitude','latitude direction','longitude','longitude direction']
```

```python
def geo_splitter(s):
    lat,long=s.split(',')
    return pd.Series([lat[:-3],lat[-1],long[:-3],long[-1]])
```

```python
geo_1=cities.Geolocation.apply(geo_splitter)
geo_1.columns=cols
```

```python
geo_1
```

```python
# utilizing handy .str() method
```

```python
geo_2=cities.Geolocation.str.split(pat='. ',expand=True)
geo_2.columns=cols
```

```python
geo_2
```
