```python
import pandas as pd
import numpy as np
```

```python
flights=pd.read_csv('data/flights.csv')
flights.head()
```

```python
# replicating pivot_table() operation with groupby() and unstack()
```

```python
flights_pivoted=flights.pivot_table(values='CANCELLED',index='AIRLINE',columns='ORG_AIR',aggfunc=np.sum,fill_value=0).astype(int).head()
flights_pivoted
```

```python
flights_agg=flights.groupby(['AIRLINE','ORG_AIR'])['CANCELLED'].sum().unstack().fillna(0).astype(int).head()
```

```python
flights_agg
```

```python
# equivalence of methods
```

```python
flights_pivoted.equals(flights_agg)
```

```python
# much more complex pivot_table()
```

```python
flights.pivot_table(index=['AIRLINE','MONTH']
                    ,columns=['ORG_AIR','CANCELLED'],values=['DEP_DELAY','DIST']
                    ,aggfunc=[np.mean,np.sum],fill_value=0).head(11).round(2)
```

```python
# replication with groupby() and unstack()
```

```python
flights.groupby(['AIRLINE','MONTH','ORG_AIR','CANCELLED'])['DEP_DELAY','DIST'].agg([np.mean,np.sum]).unstack(['ORG_AIR','CANCELLED']).swaplevel(0,1,axis=1).fillna(0).head(11).round(2)
```
