```python
# finding most outlying records with own customized function
```

```python
college=pd.read_csv('data/college.csv')
```

```python
college.head()
```

```python
def maxDeviation(s):
    z_score=(s-s.mean())/s.std()
    return z_score.abs().max()
```

```python
college.groupby('STABBR')['UGDS'].agg(maxDeviation).round(2).to_frame().head()
```

```python
college.groupby(['STABBR','RELAFFIL'])['UGDS','SATVRMID','SATMTMID'].agg([maxDeviation,np.mean,np.std]).round(2).head().style.highlight_null()
```

```python
# inserting the same function multiple times in .agg()
```

```python
grouped=college.groupby(['STABBR','RELAFFIL'])
```

```python
def pct_between(s,low,high):
    return s.between(low,high).mean()
```

```python
grouped['UGDS'].agg(pct_between,1000,3000).reset_index().head()
```

```python
try:
    grouped['UGDS'].agg([np.sum,pct_between],1000,3000).reset_index().head()
except TypeError as e:
    print(f"Error has occured: {e}")
```

```python
# wrapper for same functions and with different arguments
```

```python
def pct_between_(functionToWrap,functionName,*args,**kwargs):
    def wrapper(Series):
        return functionToWrap(Series,*args,**kwargs)
    wrapper.__name__=functionName
    return wrapper
```

```python
pct_1k_5k=pct_between_(pct_between,"fpct_1k_5k",1000,5000)
pct_3k_6k=pct_between_(pct_between,"fpct_3k_6k",3000,6000)
```

```python
grouped['UGDS'].agg([pct_1k_5k,pct_3k_6k]).head(10).style.highlight_max(color="green",axis=1)
```
