```python
import pandas as pd
import numpy as np
```

```python
college=pd.read_csv('data/college.csv',index_col='INSTNM')
```

```python
college.head()
```

```python
college.shape[0]
```

```python
grouped=college.groupby('STABBR')
grouped.ngroups == len(list(grouped.groups.keys())) == college.STABBR.nunique()
```

```python
# checking minority majority occurance with .filter() method
```

```python
def check_minority(df,threshold=.5):
    minority_pct=1-df.UGDS_WHITE
    total_minority=(df.UGDS*minority_pct).sum()
    total_ugds=df.UGDS.sum()
    total_minority_pct=total_minority/total_ugds
    return total_minority_pct > threshold
```

```python
# as seen below, new table shape differs significantly, comparing to origin size
```

```python
college_filtered=grouped.filter(check_minority)
college_filtered.shape[0]
```

```python
# equivalent method with .apply() and boolean indexing
```

```python
def return_pct_minority(df):
    minority_pct=1-df.UGDS_WHITE
    total_minority=(df.UGDS*minority_pct).sum()
    total_ugds=df.UGDS.sum()
    return total_minority/total_ugds
```

```python
pct_apply=grouped.apply(return_pct_minority)
```

```python
["%s : %.2f" % (k,v) for k,v in pct_apply.items()][:10]
```

```python
college['MINOR_MAJOR']=college.STABBR.map(pct_apply.to_dict())
```

```python
college.head()
```

```python
college.query('MINOR_MAJOR > 0.5').shape[0] == college_filtered.shape[0]
```
