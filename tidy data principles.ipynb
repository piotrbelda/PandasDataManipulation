```python
movie=pd.read_csv('data/movie.csv')
movie.head()
```

```python
actor = movie[['movie_title', 'actor_1_name','actor_2_name', 'actor_3_name',
'actor_1_facebook_likes','actor_2_facebook_likes','actor_3_facebook_likes']]
```

```python
actor.head()
```

```python
# tidying data with pd.wide_to_long() functionality
```

```python
def change_col_name(col_name):
    col_name = col_name.replace('_name', '')
    if 'facebook' in col_name:
        fb_idx = col_name.find('facebook')
        col_name = col_name[:5] + col_name[fb_idx - 1:] + col_name[5:fb_idx-1]
    return col_name
```

```python
actor2=actor.rename(columns=change_col_name)
actor2.head()
```

```python
stubs=['actor','actor_facebook_likes']
actor2_tidy=pd.wide_to_long(actor2,stubnames=stubs,i=['movie_title'],j='actor_num',sep='_')
```

```python
# tidying data with 2 stacked columns
```

```python
actor2_tidy.head()
```

```python
actor2_tidy.index.get_level_values(1).unique()
```

```python
# inverting stacked data
```

```python
usecol_func=lambda x: 'UGDS_' in x or x == 'INSTNM'
```

```python
college=pd.read_csv('data/college.csv',usecols=usecol_func,index_col='INSTNM')
college.head()
```

```python
# stacking <-> unstacking operation
```

```python
college_stacked=college.stack()
college_stacked.head(9)
```

```python
college_stacked.unstack().head()
```

```python
college2=pd.read_csv('data/college.csv',usecols=usecol_func)
college2.head()
```

```python
college_melted=college2.melt(id_vars=['INSTNM'],var_name='RACE',value_name='PERC')
college_melted.head()
```

```python
# heading back from melted formula to origin with .pivot() method
```

```python
melted_inv=college_melted.pivot('INSTNM','RACE','PERC')
melted_inv.head()
```

```python
# unstacking after a groupby operation
```

```python
employee=pd.read_csv('data/employee.csv')
```

```python
agg=employee.groupby(['RACE','GENDER'])['BASE_SALARY'].agg(['mean','max','min']).astype(int)
agg.unstack(1)
```

```python
agg.unstack(1).equals(agg.unstack('GENDER'))
```
